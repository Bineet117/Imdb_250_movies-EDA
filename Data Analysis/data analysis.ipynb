## Data Visualization

### 1. Movies of each category

```python
# plt.figure(figsize=(7, 4))
ax = sns.countplot(x = "Genre" , data = df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.ylabel('No. of Movies')
plt.xticks(rotation=45)


plt.show()
```

From the above graph we can say that most of the movies are of `drama` category 

### 2. Top 5 director whose movies has crossed a metascore of 85

```python
df[df["Metascore"] > 85]["Director"].value_counts().head()
```

### 3. Top 5 director whose movies has crossed a rating of 8.0

```python
df[df["Rating"] > 8.0]["Director"].value_counts().head()

# plt.bar(["Christopher Nolan","Steven Spielberg","Stanley Kubrick","Martin Scorsese","Alfred Hitchcock"],df[df["Rating"] > 8.0]["Director"].value_counts().head() )
# plt.xticks(rotation = 45)
```

### 4. Which decade has most movies

```python
plt.figure(figsize = (5,3))
sns.histplot(df["Year"]  , kde = True )
```

from the above graph (2000-2010) has most no. of movies

### 5. Top 5 movies category whose rating is more than 8.5 

```python
df[df["Rating"] > 8.5]["Genre"].value_counts().head(5)
```

### 6. Top 5 oldest movies

```python
for i in df.sort_values(by = "Year" ).head()["Movie"].values:
    print(i)
```

### 7. Runtime Analysis
There is a column named `Runtime` in the dataframe which primarily shows the length of the movie. It might be intersting to see how this variable this distributed. Plot a `histogram` or `distplot` of seaborn to find the Runtime range most of the movies fall into.

```python
plt.figure(figsize = [5,3])
runtime=sns.histplot(data = df , x = df["Time"] , kde = True )

plt.ylabel('No. of Movies')

```

Most of the movies appear to be sharply 2 hour-long.

### 8. Movie Release Years by Genre: A Boxplot Visualization

```python
sns.boxplot(x= "Genre",  y = "Year" , data = df)
plt.xticks(rotation=45)
plt.show()
```

### 9. Number of movies by each director

```python
plt.figure(figsize=(30, 10))
sns.countplot(x  = "Director" , data = df)

plt.xlabel('Director',fontsize=30)
plt.ylabel('No. of movies',fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

most of the director has 1 movie

### 10. Ratio of each categories in `Genre`

```python
# df["Genre"].value_counts(normalize = True)
plt.figure(figsize = (4,4))
num = df["Genre"].value_counts().head(9)
plt.pie(num , autopct = "%.2f",labels =num.index )
plt.show()
```
